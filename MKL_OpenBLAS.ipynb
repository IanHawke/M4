```julia
using PyPlot
```

```julia
n=[5,10,20,40,80,160,320,750,1000,2000,5000,10000]
```

```julia
tMKL=[.000073,.00011,.0002,.000691,.004720,.0235,.083434,.5,1.042934,6.01,62.47,456.623]
```

```julia
tOB=[.000057,.000149,.00038,.002358,.015,.07,.25,1.5,2.55,12.5,NaN,NaN]
```

```julia
loglog(n,tMKL,label="MKL",marker="o")
loglog(n,tOB,label="OpenBLAS",marker="o")
legend(loc="upper left")
xlabel("Rank of Random Matrix")
ylabel("Time for eigfact (s)")
title("Time Scaling for MKL vs. OpenBLAS")
```

```julia
semilogx(n,tOB./tMKL,marker="o")
xlabel("Rank of Random Matrix")
ylabel("OpenBLAS/ MKL")
title("Speed up from MKL")

annotate("OpenBLAS aborting",
xy=[5e3;2],
xycoords="data",
xytext=[.75e2;2],
arrowprops=["facecolor"=>"black"])
```

```julia
mem=[9,16,49,158,549,2000,7000,42000,76000,328000,1858000,7430000]
```

```julia
loglog(n,(mem[end]/n[end]^2)*n.^2,label="Quadratic Scaling")
scatter(n,mem,marker="o",color="red",label="Data")
xlabel("Rank of Random Matrix")
ylabel("Memory Allocated (kb)" )
title("Memory Scaling")
legend(loc="upper left")
```
