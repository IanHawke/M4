```julia
f(x,y)=sin.(x.+y)
```

```julia
using Plots
using Calculus
pl()
```

```julia
xs=range(-π,stop=π,length=100)
ys=range(-π,stop=π,length=100)

xa=repeat(xs,1,length(ys))
ya=repeat(transpose(ys),length(xs),1)
za=f(xa,ya);
```

```julia
surface(xs,ys,za)
surface!(xs,ys,zeros(Float64,length(xs),length(ys)),alpha=0.5)
```
