## The DFT

$$
\sum_{i=0}^{N-1} e^{i k j/N} x_j
$$

```julia
function Fk(k::Float64,y::Array)
    N=length(y)
    fi=0
    for jj in 1:N
        fi+=exp(-im*k*jj/N)*y[jj]
    end
    return fi
end
```

```julia
xmax=10π
xmin=0
N=2^3

ldomain=xmax-xmin;

x=range(xmin,stop=xmax,length=N)
y=sin.(x);
```

```julia
k1=2π*collect(0:N);
k2=2π*collect(0:.2:N);
```

```julia
DFT1=zeros(Complex{Float64},length(k1))
for ii in 1:length(k1)
    DFT1[ii]=Fk(k1[ii],y)
end
DFT2=zeros(Complex{Float64},length(k2))
for ii in 1:length(k2)
    DFT2[ii]=Fk(k2[ii],y)
end
```

```julia
scatter(k1,abs.(DFT1),markersize=10,xlims=[0,50])
scatter!(k2,abs.(DFT2))
```
