\begin{equation}
    f= \frac{N_{circle}}{N_{square}} = \frac{\pi r^2}{4 r^2}= \frac{\pi}{4}
\end{equation}
\begin{equation}
    \pi = 4 f
\end{equation}

```julia
using PyPlot
```

```julia
function lt(r)
    if r<.5
        return true
    else
        return false
    end
end
```

```julia
N=[10,50,100,500,1000,5000,10000];
M=15;
```

```julia
πapprox=zeros(Float64,length(N),M);

for ii in 1:length(N)
    for jj in 1:M
        X=zeros(Float64,N[ii]);
        Y=zeros(Float64,N[ii]);
        rand!(X);
        rand!(Y);
        R=sqrt((X-.5).^2+(Y-.5).^2);
        πapprox[ii,jj]=4*length(filter(lt,R))/N[ii];
    end
end
```

```julia
πave=sum(πapprox,2)/M;
πstd=sum((πapprox-repmat(πave,1,M)).^2,2)/M;
```

```julia
title("Monte Carlo Estimation of π")
ylabel("π estimate")
xlabel("Log N points")
plot(log(N),π*ones(N))
plot(log(N),πapprox,marker="o")
```

```julia
title("Dependence of Monte Carlo Error on Number of Points")
ylabel("standard deviation")
xlabel("Log N points")
plot(log10(N),πstd,marker="o")
```

```julia
plot(log(N),π*ones(N))
plot(log(N),πave,marker="o")
```

```julia
X=zeros(Float64,1000);
Y=zeros(Float64,1000);
rand!(X);
rand!(Y);
R=sqrt((X-.5).^2+(Y-.5).^2);
Xc=[];
Yc=[]
for ii in 1:length(X)
    if R[ii]<.5
        push!(Xc,X[ii]);
        push!(Yc,Y[ii]);
    end
end

title("The dartboard")
xlim(0,1)
ylim(0,1)
scatter(X,Y)
scatter(Xc,Yc,color="red")    
```
